### Importing Necessary Library and doing some basic checks on data 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Expanded_data_with_more_features.csv")
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.info()
```

### Removing null values

```python
df.isnull().sum()
```

```python
newdf=df.dropna()
```

```python
newdf.isnull().sum()
```

```python
newdf.head(10)
```

```python
newdf=newdf.drop(['Unnamed: 0'],axis=1)
```

```python
newdf.head(10)
```

### Gender distribution

```python
plt.figure(figsize=(8,5))
s=sns.countplot(newdf,x='Gender')
s.bar_label(s.containers[0])
plt.show()
```

### Students score by parents education

```python
gb = newdf.groupby('ParentEduc').agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
plt.figure(figsize=(8,5))
sns.heatmap(gb,annot=True)
plt.title("Students score by parents education")
plt.show()
```

### Marital Status of Parents

```python
plt.figure(figsize=(8,5))
s=sns.countplot(newdf,x='ParentMaritalStatus')
s.bar_label(s.containers[0])
plt.show()
```

### Impact on student by parents marital satus

```python
gb = newdf.groupby('ParentMaritalStatus').agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
plt.figure(figsize=(8,5))
sns.heatmap(gb,annot=True)
plt.title("Impact on student by parents marital satus")
plt.show()
```

### Students those who play Sport

```python
sports = newdf['PracticeSport'].groupby(newdf['ParentMaritalStatus']).value_counts().unstack()

ax = sports.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Parent Marital Status')
plt.ylabel('Count')
plt.title('Students those who play Sport')
plt.legend(title='Practice Sport')

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

plt.show()
```

### Students those who play Sport by gender

```python
sports = newdf['Gender'].groupby([newdf['PracticeSport']]).value_counts().unstack()

ax = sports.plot(kind='bar', figsize=(10, 6))

plt.ylabel('Count')
plt.title('Practice Sport by Gender')
plt.legend(title='Gender')

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

plt.show()
```

### Student who play sports on the basis of Parent Marital Status

```python
sports = newdf['ParentMaritalStatus'].groupby([newdf['Gender'], newdf['PracticeSport']]).size().unstack()

fig, ax = plt.subplots(figsize=(15, 10))
sports.plot(kind='bar', ax=ax)

plt.ylabel('Count')
plt.title('Practice Sport by Gender and Parent Marital Status')
plt.legend(title='Marital Status')
plt.xticks(rotation=10, ha='right')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=newdf,x='EthnicGroup')
plt.bar_label(sns.countplot(data=newdf,x='EthnicGroup').containers[0])
```

```python
sports = newdf['EthnicGroup'].groupby([newdf['Gender'], newdf['WklyStudyHours']]).value_counts().unstack()

fig, ax = plt.subplots(figsize=(15, 10))
sports.plot(kind='bar', ax=ax)

plt.xlabel('Study Hours')
plt.ylabel('Count')
plt.title('Ethnic Group Study Hour by Gender')
plt.legend(title='Study Hours')
plt.xticks(rotation=10, ha='right')

plt.tight_layout()
plt.show()
```
